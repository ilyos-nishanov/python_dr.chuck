```python
#old way find()

hand = open('mbox-short.txt')
for line in hand:
    line = line.rstrip()
    if line.find('From:') >= 0:
        print(line)
```

```python
#old way: startswith()
hand = open('mbox-short.txt')
for line in hand:
    line = line.rstrip()
    if line.startswith('From:'):
        print(line)
```

```python
# with regex
import re
hand = open('mbox-short.txt')
for line in hand:
    line = line.rstrip()
    if re.search('^From:',line):
        print(line)
```

```python
#findall()
x = 'My 2 favorite numbers are 19 and 42'
y = re.findall('[0-9]+',x)
y
```

```python
x = 'From: Using the : character'
y = re.findall('^F.+:',x)
y
```

```python
x = 'From: Using the : character'
y = re.findall('^F.+?:',x)
y
```

```python
fhandle = open('mbox-short.txt')
for line in fhandle:
    y = re.findall('\S+@\S+',line)
    print(y)
```
